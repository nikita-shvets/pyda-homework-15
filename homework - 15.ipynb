```python
import pandas as pd
dataset = pd.read_csv('visit_log.csv',sep = ';')
def source(data):
    if data['traffic_source'] == 'yandex' or data['traffic_source'] == 'google':
        return 'organic'
    if data['traffic_source'] == 'paid' or data['traffic_source'] == 'email':
        if data['region'] == 'Russia':
            return 'ad'
        return 'other'
    return data['traffic_source']
dataset['source_type'] = dataset.apply(source, axis = 1)
dataset
```
